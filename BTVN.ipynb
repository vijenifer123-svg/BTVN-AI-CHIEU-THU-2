```python
#bài 23.7
from osmnx.routing import route_to_gdf
!pip install folium geopy osmnx networkx
import osmnx as ox
import networkx as nx
import folium
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
a=Nominatim(user_agent='leo')
loc_A=a.geocode("Ben Thanh Market, Ho Chi Minh City, Vietnam")
loc_B=a.geocode("Tan Son Nhat Airport, Ho Chi Minh City, Vietnam")
lat_A = (loc_A.latitude, loc_A.longitude)
lat_B = (loc_B.latitude, loc_B.longitude)
midpoint=((lat_A[0]+lat_B[0])/2,(lat_A[1]+lat_B[1])/2)
G=ox.graph_from_point(midpoint,dist=3000,network_type='drive')
dau=ox.distance.nearest_nodes(G,lat_A[1],lat_A[0])
cuoi=ox.distance.nearest_nodes(G,lat_B[1],lat_B[0])
#PHƯƠNG PHÁP 1: DIJKSTRA
route=nx.shortest_path(G,dau,cuoi,weight='length')
list_node=[ ]
for n in route:
     latitude=G.nodes[n]['y']
     longitude=G.nodes[n]['x']
     list_node.append((latitude,longitude))

#PHƯƠNG PHÁP 2: A*
def heuristic(u, v):
    node_u = G.nodes[u]
    node_v = G.nodes[v]
    return geodesic((node_u['y'], node_u['x']), (node_v['y'], node_v['x'])).m
route_astar = nx.astar_path(G, dau, cuoi, heuristic=heuristic, weight='length')
list_node_astar=[ ]
for n in route_astar:
     latitude=G.nodes[n]['y']
     longitude=G.nodes[n]['x']
     list_node_astar.append((latitude,longitude))


m = folium.Map(location=midpoint, zoom_start=14)
folium.Marker(lat_A, popup="Điểm A", icon=folium.Icon(color='red')).add_to(m)
folium.Marker(lat_B, popup="Điểm B", icon=folium.Icon(color='blue')).add_to(m)
folium.PolyLine(list_node, color='green', weight=3, opacity=0.5).add_to(m)
folium.PolyLine(list_node_astar, color='red', weight=8, opacity=0.4).add_to(m)
display(m)

Thuật toán Dijkstra (nx.shortest_path)
Nguyên lý hoạt động: Thuật toán duyệt mở rộng dần đều từ điểm bắt đầu ra mọi hướng (như vết dầu loang) cho đến khi chạm tới đích.
Ưu điểm:
Độ tin cậy tuyệt đối: Luôn đảm bảo chắc chắn tìm ra được con đường có tổng trọng số (chiều dài) ngắn nhất.
Khám phá toàn diện: Rất phù hợp nếu hệ thống muốn tìm đường đi từ 1 kho hàng đến tất cả các khách hàng xung quanh.
Nhược điểm:
Hiệu suất thấp: Vì nó lan tỏa ra mọi hướng không có chủ đích, nó sẽ lãng phí tài nguyên máy tính để dò dẫm những con đường đi ngược hướng với đích đến. Trong một bản đồ đô thị lớn (như toàn TP.HCM), thuật toán sẽ chạy rất chậm.
2. Thuật toán A (A-Star) (nx.astar_path)*
Nguyên lý hoạt động: Kế thừa Dijkstra nhưng thông minh hơn nhờ được trang bị thêm một "hàm đánh giá" (Heuristic function - trong code trên là ước lượng khoảng cách đường chim bay đến đích).
Ưu điểm:
Định hướng và Tốc độ cao: Thay vì lan tỏa mù quáng ra mọi hướng, A* luôn ưu tiên duyệt những con đường có xu hướng tiến gần về phía đích. Nhờ đó, số lượng điểm phải kiểm tra (nodes) ít hơn rất nhiều, giúp hệ thống tính toán lộ trình chớp nhoáng (đặc biệt quan trọng đối với các app gọi xe công nghệ hay điều phối shipper theo thời gian thực).
Nhược điểm:
Phụ thuộc vào hàm Heuristic: Nếu bạn thiết kế hàm Heuristic tồi (đánh giá sai khoảng cách), A* có thể rơi vào bẫy và chỉ ra một con đường chưa phải là tối ưu nhất. Thêm vào đó, việc tính toán thêm hàm Heuristic ở mỗi bước cũng đòi hỏi cấu trúc code phức tạp hơn một chút.



```

```python
#bài 23.8
!pip install folium geopy osmnx networkx
import osmnx as ox
import networkx as nx
import folium
place = "District 1, Ho Chi Minh City, Vietnam"
G = ox.graph_from_place(place, network_type='drive')
m = folium.Map(location=[10.775, 106.700], zoom_start=13)
customers = [
    ("Khách A (Chợ Bến Thành)", (10.7725, 106.6980)),
    ("Khách B (Phố đi bộ)", (10.7745, 106.7030))
]

drivers = [
    ("Xe Grab 1", (10.7760, 106.7010)),
    ("Xe Grab 2", (10.7710, 106.6950)),
    ("Xe Grab 3", (10.7800, 106.6900))
]
cust_node=[]
for name,coord in customers:
    node_gannhat=ox.distance.nearest_nodes(G,coord[1],coord[0])
    cust_node.append((name,node_gannhat))
    folium.Marker(coord,popup=name,icon=folium.Icon(color='red', icon='user')).add_to(m)
driver_node=[]
for name,coord in drivers:
    node_gan=ox.distance.nearest_nodes(G,coord[1],coord[0])
    driver_node.append((name,node_gan))
    folium.Marker(coord, popup=name, icon=folium.Icon(color='blue', icon='car')).add_to(m)

bang_xep_hang_xe = []
mau_sac = ['green', 'purple', 'orange']
chi_so_mau = 0
for khach in cust_node:
    ten_khach = khach[0]
    nut_cua_khach = khach[1]

    xe_da_nhan_khach = []

    for xe in driver_node:
        ten_xe = xe[0]
        nut_cua_xe = xe[1]

        if ten_xe not in xe_da_nhan_khach:
            if nx.has_path(G, nut_cua_xe, nut_cua_khach):
                duong_di = nx.shortest_path(G, nut_cua_xe, nut_cua_khach, weight='length')
                khoang_cach = nx.path_weight(G, duong_di, weight='length')
                bang_xep_hang_xe.append((khoang_cach, ten_xe, duong_di))
    if len(bang_xep_hang_xe) > 0:
                bang_xep_hang_xe.sort()
                chuyen_di_tot_nhat = bang_xep_hang_xe[0]
                khoang_cach_ngan_nhat = chuyen_di_tot_nhat[0]
                xe_duoc_chon = chuyen_di_tot_nhat[1]
                duong_di_chinh_thuc = chuyen_di_tot_nhat[2]
                xe_da_nhan_khach.append(xe_duoc_chon)
    list_node=[ ]
    for n in duong_di_chinh_thuc:
        latitude=G.nodes[n]['y']
        longitude=G.nodes[n]['x']
        list_node.append((latitude,longitude))
    folium.PolyLine(
                list_node,
                color=mau_sac[chi_so_mau % len(mau_sac)],
                weight=5,
                opacity=0.8,
                tooltip=f"{xe_duoc_chon} đang đón {ten_khach}"
            ).add_to(m)

    chi_so_mau += 1
m
```

```python
#bai 23.15
!pip install folium pandas scikit-learn numpy gradio

# 2. IMPORT THƯ VIỆN
import numpy as np
import pandas as pd
import folium
import gradio as gr
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# ==========================================
# PHẦN 1: HUẤN LUYỆN AI (TRAINING PHASE)
# ==========================================
print("Đang huấn luyện các mô hình AI...")

# Tạo dữ liệu lịch sử giả lập (Historical Data)
np.random.seed(42)
num_historical_orders = 500
center_lat, center_lon = 10.7769, 106.7009 # Trung tâm TP.HCM
hist_lats = center_lat + np.random.normal(0, 0.05, num_historical_orders)
hist_lons = center_lon + np.random.normal(0, 0.05, num_historical_orders)
hist_weights = np.random.uniform(1, 20, num_historical_orders)

X_hist = np.column_stack((hist_lats, hist_lons))

# AI Model 1: K-Means (Học không giám sát để quy hoạch 5 Hubs trung chuyển)
ai_router = KMeans(n_clusters=5, random_state=42, n_init=10)
hub_assignments = ai_router.fit_predict(X_hist)
hubs_locations = ai_router.cluster_centers_

# Tính khoảng cách từ mỗi đơn hàng lịch sử đến Hub của nó để làm features
distances = np.linalg.norm(X_hist - hubs_locations[hub_assignments], axis=1)

# Tạo dữ liệu thời gian giao hàng (ETA) giả lập: Phụ thuộc khoảng cách, khối lượng và nhiễu (kẹt xe)
# Thời gian (phút) = khoảng cách * 1500 + khối lượng * 0.5 + nhiễu ngẫu nhiên
eta_historical = (distances * 1500) + (hist_weights * 0.5) + np.random.normal(5, 2, num_historical_orders)

# AI Model 2: Random Forest Regressor (Học có giám sát để dự đoán Thời gian giao hàng)
X_eta = np.column_stack((distances, hist_weights))
ai_eta_predictor = RandomForestRegressor(n_estimators=100, random_state=42)
ai_eta_predictor.fit(X_eta, eta_historical)

print("Hoàn tất huấn luyện AI!")

# ==========================================
# PHẦN 2: XÂY DỰNG LOGIC SUY LUẬN (INFERENCE)
# ==========================================
def ai_dispatch_order(new_lat, new_lon, weight):
    """Hàm này là 'Trái tim' của ứng dụng, nhận lệnh mới và dùng AI để xử lý"""
    new_coords = np.array([[new_lat, new_lon]])

    # 1. AI Phân loại Hub tối ưu cho đơn hàng mới
    assigned_hub_id = ai_router.predict(new_coords)[0]
    hub_loc = hubs_locations[assigned_hub_id]

    # 2. Tính toán khoảng cách tuyến đường
    route_distance = np.linalg.norm(new_coords[0] - hub_loc)

    # 3. AI Dự đoán thời gian giao hàng (ETA)
    features_for_eta = np.array([[route_distance, weight]])
    predicted_eta = ai_eta_predictor.predict(features_for_eta)[0]

    # 4. Trực quan hóa bản đồ cho quyết định của AI
    # Khởi tạo bản đồ lấy tâm là đơn hàng mới
    m = folium.Map(location=[new_lat, new_lon], zoom_start=13, tiles='CartoDB positron')

    # Vẽ toàn bộ 5 Hubs
    for i, loc in enumerate(hubs_locations):
        color = 'red' if i == assigned_hub_id else 'gray'
        icon_type = 'star' if i == assigned_hub_id else 'home'
        folium.Marker(
            location=[loc[0], loc[1]],
            popup=f"Hub {i}",
            icon=folium.Icon(color=color, icon=icon_type, prefix='fa')
        ).add_to(m)

    # Vẽ điểm giao hàng mới
    folium.Marker(
        location=[new_lat, new_lon],
        popup=f"New Order ({weight}kg)",
        icon=folium.Icon(color='blue', icon='box', prefix='fa')
    ).add_to(m)

    # Vẽ đường nối (Routing) từ Hub tối ưu đến điểm giao
    folium.PolyLine(
        locations=[hub_loc, [new_lat, new_lon]],
        color='blue',
        weight=4,
        opacity=0.7,
        tooltip=f"Tuyến đường AI đề xuất"
    ).add_to(m)

    # Xuất bản đồ ra định dạng HTML để hiển thị trên UI
    map_html = m._repr_html_()

    # Thông báo kết quả từ AI
    result_text = (
        f"✅ AI ĐÃ XỬ LÝ ĐƠN HÀNG:\n"
        f" - Tọa độ: {new_lat:.4f}, {new_lon:.4f}\n"
        f" - Khối lượng: {weight} kg\n"
        f"🎯 Quyết định:\n"
        f" - Gán cho: Kho trung chuyển (Hub) số {assigned_hub_id}\n"
        f" - Thời gian giao hàng dự kiến (ETA): {predicted_eta:.1f} phút"
    )

    return map_html, result_text

# ==========================================
# PHẦN 3: XÂY DỰNG GIAO DIỆN APP (GRADIO UI)
# ==========================================
# Tạo giao diện Web App nhập liệu
with gr.Blocks(theme=gr.themes.Soft()) as logistics_app:
    gr.Markdown("# 🚚 Ứng dụng AI Điều phối Logistics (Smart Dispatch App)")
    gr.Markdown("Nhập tọa độ và khối lượng đơn hàng mới. AI sẽ tự động tính toán Hub phù hợp nhất để xuất kho và dự báo thời gian giao hàng.")

    with gr.Row():
        with gr.Column():
            lat_input = gr.Slider(minimum=10.60, maximum=10.90, value=10.78, label="Vĩ độ (Latitude - TP.HCM)")
            lon_input = gr.Slider(minimum=106.50, maximum=106.90, value=106.70, label="Kinh độ (Longitude - TP.HCM)")
            weight_input = gr.Number(value=5.0, label="Khối lượng đơn hàng (kg)")
            submit_btn = gr.Button("🚀 Yêu cầu AI Điều phối", variant="primary")

        with gr.Column():
            text_output = gr.Textbox(label="Báo cáo quyết định của AI", lines=6)

    # Hiển thị bản đồ
    map_output = gr.HTML(label="Bản đồ lộ trình")

    # Kết nối nút bấm với hàm xử lý
    submit_btn.click(
        fn=ai_dispatch_order,
        inputs=[lat_input, lon_input, weight_input],
        outputs=[map_output, text_output]
    )

# Chạy ứng dụng trên Colab
logistics_app.launch(debug=True)
```
